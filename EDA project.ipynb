```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data(filepath,file_type):
    if file_type == 'csv':
        return pd.read_csv(filepath)
    elif file_type == 'excel':
        return pd.read_excel(filepath)
    else:
        raise ValueError("Unsupported file type")
```

```python
class visualize_data:
    def visualize_numarical_values(df):
        # Visualize numerical data in histogram
        num_columns = df.select_dtypes(include=['int64','float64']).columns
        for col in num_columns:
            #Schema for ploting: Size, type, title, xlabel, ylabel, grid,show (STTXYGS)
            plt.figure(figsize=(10,4))
            df[col].hist(bins=50)
            plt.title(f"Histogram of {col}'s")
            plt.xlabel(col)
            plt.ylabel('Frequency')
            plt.show()
    def visualize_categorical_data(df):
        # Visualize categorical data in bars
        cat_columns = df.select_dtypes(include=['object','category']).columns
        for col in cat_columns:
            #Schema for ploting: Size, type, title, xlabel, ylabel, grid,show (STTXYGS)
            value_counts = df[col].value_counts()
            plt.figure(figsize=(15,4))
            value_counts = value_counts.iloc[:20] # I used this here as when I don't use it, the data in the x-axis won't be readable
            value_counts.plot(kind='bar')
            plt.title(f'Bar Chart of {col}')
            plt.xlabel(col)
            plt.ylabel('Frequency')
            plt.grid(False)
            plt.show()
    def visualize_date_columns(df):
        df['data_dte'] = pd.to_datetime(df['data_dte'])
        #Schema for ploting: Size, type, title, xlabel, ylabel, grid,show (STTXYGS)
        df.sort_values(by='data_dte', inplace=True)
        plt.figure(figsize=(12, 6))
        plt.plot(df['data_dte'], df['Total'], marker='o', linestyle='-')
        plt.title('Total Over Time')
        plt.xlabel('Date')
        plt.ylabel('Total')
        plt.show()
    def visualize_years(df):
        df['Year'] = df['Year'].sort_values().values
        year_counts = df['Year'].value_counts()
        plt.figure(figsize=(10,10))
        plt.pie(year_counts, labels=year_counts.index.values.tolist())
        plt.show()
    def visualize_scatter(df):
        plt.figure(figsize=(10, 6))
        plt.scatter(df['Year'], df['Total'])
        plt.title('Scatter Plot of Total by Year')
        plt.xlabel('Year')
        plt.ylabel('Total')
        plt.show()
```
